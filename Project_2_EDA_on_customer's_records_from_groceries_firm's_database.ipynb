How many total amount spent overall on fruits , meat , fish , sweet and gold also give in percentage wise ?

```python
overall_amt_Fruits = cstm_per_df.MntFruits.sum()
overall_amt_MeatProducts = cstm_per_df.MntMeatProducts.sum()
overall_amt_FishProducts = cstm_per_df.MntFishProducts.sum()
overall_amt_MntSweetProducts = cstm_per_df.MntSweetProducts.sum()
overall_amt_Gold = cstm_per_df.MntGoldProds.sum()
```

```python
overall_total_amt = cstm_per_df.Total_amount_spend.sum()
pert_for_Fruit_Prd = (cstm_per_df.MntFruits.sum() / cstm_per_df.Total_amount_spend.sum()) * 100
pert_for_Meat_Prd = (cstm_per_df.MntMeatProducts.sum() / cstm_per_df.Total_amount_spend.sum()) * 100
pert_for_Fish_Prd = (cstm_per_df.MntFishProducts.sum() / cstm_per_df.Total_amount_spend.sum()) * 100
pert_for_Sweet_Prd = (cstm_per_df.MntSweetProducts.sum() / cstm_per_df.Total_amount_spend.sum()) * 100
pert_for_Gold_Prd = (cstm_per_df.MntGoldProds.sum() / cstm_per_df.Total_amount_spend.sum()) * 100
```

```python
print( 'The Total Amount Which Spend on Fruit is {} Rs and the Percentage for that amount is {:.2f}%.'.format(overall_amt_Fruits,pert_for_Fruit_Prd))
print( 'The Total Amount Which Spend on Meat is {} Rs and the Percentage for that amount is {:.2f}%.'.format(overall_amt_MeatProducts,pert_for_Meat_Prd))
print( 'The Total Amount Which Spend on Fish is {} Rs and the Percentage for that amount is {:.2f}%.'.format(overall_amt_FishProducts,pert_for_Fish_Prd))
print( 'The Total Amount Which Spend on Sweet is {} Rs and the Percentage for that amount is {:.2f}%.'.format(overall_amt_MntSweetProducts,pert_for_Sweet_Prd))
print( 'The Total Amount Which Spend on Gold is {} Rs and the Percentage for that amount is {:.2f}%.'.format(overall_amt_Gold,pert_for_Gold_Prd))
```
